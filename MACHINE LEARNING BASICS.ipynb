```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer_df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
cancer_df.head()
```

```python
X=cancer_df
y=cancer['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
pred1=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,pred1))
print('\n')
print(confusion_matrix(y_test,pred1))
print('\n')
print(accuracy_score(y_test,pred1))
```

```python
#TO IMPROVE OVERALL ACCURACY WE USE GRIDSEARCH CV
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
grid_params={'C':[0.1,1,10,100,1000,2000],'gamma':[10,1,0.1,0,0.01,0.001]}
```

```python
grid=GridSearchCV(SVC(),grid_params,verbose=2)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
pred2=grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,pred2))
print('\n')
print(confusion_matrix(y_test,pred2))
print('\n')
print(accuracy_score(y_test,pred2))
```
